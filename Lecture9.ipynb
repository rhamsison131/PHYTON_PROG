```python
X_data, real_groups = make_blobs(
    n_samples = 300,
    centers   = 3,
    cluster_std = 1.0,
    random_state = 99
)

dataset = pd.DataFrame(X_data, columns=["FeatA","FeatB"])
dataset["OriginalLabel"] = real_groups
dataset.head()
```

```python
# Import KMeans from sklearn.cluster
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np

# Create or load your data
# For example:
# Option 1: Create sample data
data = pd.DataFrame({
    'feature1': np.random.rand(100),
    'feature2': np.random.rand(100)
})

# Option 2: Load data from a file
# data = pd.read_csv('your_data.csv')

# Define X as the features to use for clustering
X = data[['feature1', 'feature2']]  # Replace with your actual feature columns

# Initialize KMeans
kmeans = KMeans(n_clusters=4, random_state=42)

# Fit KMeans to the data
kmeans.fit(X)

# Predict cluster labels
data['KMeansLabel'] = kmeans.predict(X)
data.head()
```

```python
# Import matplotlib.pyplot and other necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

# Create sample data with appropriate column names
# (Replace this with your actual data loading if you have a specific dataset)
np.random.seed(42)
data = pd.DataFrame({
    'Feature1': np.random.rand(100) * 10,
    'Feature2': np.random.rand(100) * 10
})

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
data['KMeansLabel'] = kmeans.fit_predict(data[['Feature1', 'Feature2']])

# Plotting the clusters
plt.figure(figsize=(8,6))
plt.scatter(data['Feature1'], data['Feature2'], c=data['KMeansLabel'], cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='red', marker='X', s=200, label='Centroids')
plt.title("KMeans Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
